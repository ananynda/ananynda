```python
llm = Llama(
    model_path=model_path,
    n_ctx=1024,  # Context window
)
```

**Utility functions**

```python
# defining a function to parse the JSON output from the model
def extract_json_data(json_str):
    try:
        # Find the indices of the opening and closing curly braces
        json_start = json_str.find('{')
        json_end = json_str.rfind('}')

        if json_start != -1 and json_end != -1:
            extracted_category = json_str[json_start:json_end + 1]  # Extract the JSON object
            data_dict = json.loads(extracted_category)
            return data_dict
        else:
            print(f"Warning: JSON object not found in response: {json_str}")
            return {}
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
        return {}
```

```python
# creating a copy of the data
data_1 = data.copy()
```

```python
#Defining the response funciton for Task 1.
def response_1(prompt,ticket):
    model_output = llm(
      f"""
      Q: {prompt}
      Support ticket: {ticket}
      A:
      """,
      max_tokens=10, #Complete the code to set the maximum number of tokens the model should generate for this task.
      stop=["Q:", "\n"],
      temperature=0.01, #Complete the code to set the value for temperature.
      echo=False,
    )

    temp_output = model_output["choices"][0]["text"]
    final_output = temp_output[temp_output.index('{'):]

    return final_output
```

```python
# Defining prompt for task 1
prompt_1 = """
   You are an AI categorizing support tickets for an IT department.
   Please categorize the support tickets in to one of the following categories: Hardware Issues, Data Recovery, or Technical Issues.
   The output category should be formatted as follows: {"category": "Hardware Issues"}, or {"category": "Data Recovery"}, or {"category": "Technical Issues"}.
   Please ensure that all curly braces are closed and there are no additional characters in the output.
"""
```

```python
# Applying generate_llama_response function on support_ticket_text column
start = time.time()
data_1['model_response'] = data_1['support_ticket_text'].apply(lambda x: response_1(prompt_1, x))
end = time.time()
```

**Actionable Insights and Recommendations**

*Insights*:

Outputs generated by the model will be more robust if the prompts are fed with information about the company in specific.
Outputs can be tailored to a specific business either in the prompts or after outputs are generated (see Business Specific Output section).
Outputs for categories can be changed/increased if it makes sense for the types of support offered by a specific business.
Outputs for priority are biased towards high priority, this could change depending on the availabilty of a specific business to provide such solutions.
Outputs for eta are either 1 or 3 days depending on how long the model expects the issue to take to be resolved - this could change depending on actual support availability of a specific business.
Technical issues are estimated to be solved the quickest, while hardware issues and data recovery are estimated to take longer to resolve. This makes sense for real world applications.*

Recommendations:

Fine tune the model for your business specifically (i.e. adding your business name to the responses).
Adjust priority of support tickets to reflect priorities the business can actually facilitate.
Adjust eta of support tickets to reflect actual eta the business needs to resolve these issues.
Determine if response column entries should contain python code such as, \n, for formatting. Depending on how the responses are sent to customers, code such as this can be removed.
Review model output for actual working data to ensure satisfactory performance before implementation.
